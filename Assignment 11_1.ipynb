```python
#Course: Data Science Masters
#Title: Pandas - Assignment 11.1            
#Tool: Jupyter notebook 
#Author: Krishna A
#Date: 22-04-2018 (DD/MM/YYYY)

#Problem Statement 
    #0 : Read the dataset from the below link
    #https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

        
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   #import matplotlib

sFile = r'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(sFile)
print(df.head(5))
```

```python
# Problem 1: Chart to plot: Create a pie chart presenting the male/female proportion_

X = [df['sex'].value_counts()['male'], df['sex'].value_counts()['female']]
labels = 'Male','Female'
plt.pie(X, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Problem 2: Chart to plot: Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

x = df['fare'].values
y = df['age'].values

colors = np.where(df['sex']=='male','blue','pink')  # Male in Blue and Female in Pink

plt.scatter(x, y, c=colors, alpha=0.5)
plt.show()
```
